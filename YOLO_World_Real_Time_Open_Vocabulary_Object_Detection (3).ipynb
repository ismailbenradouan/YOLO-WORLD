```python
#  Install libraries
!pip install -q ultralytics opencv-python-headless pandas

# Imports
from ultralytics import YOLO
from IPython.display import display, Video
from PIL import Image as PILImage
from google.colab import files
import requests
from io import BytesIO
import pandas as pd

# Load YOLO model
model = YOLO("yolov8s.pt")

while True:
    # Choose input
    choice = input("\nType 'upload', 'url', or 'exit': ").lower()
    if choice == 'exit':
        print("Exiting...")
        break
    elif choice == 'upload':
        uploaded = files.upload()
        file_path = list(uploaded.keys())[0]
    elif choice == 'url':
        url = input("Enter file URL: ")
        response = requests.get(url)
        file_path = BytesIO(response.content)
    else:
        print("Invalid choice")
        continue

    # Image or Video
    file_type = input("Is this an 'image' or 'video'? ").lower()

    # Confidence threshold with validation
    while True:
        conf_input = input("Enter confidence threshold (0 → 1): ")
        try:
            conf_thresh = float(conf_input)
            if 0 <= conf_thresh <= 1:
                break
            else:
                print("Number must be between 0 and 1")
        except ValueError:
            print("Invalid! Use a number like 0.25 or 0.5")

    if file_type == 'image':
        results = model.predict(file_path, conf=conf_thresh)[0]
        display(PILImage.open(file_path) if choice=="url" else PILImage.open(file_path))
        results.plot(show=True)

        print("\nDetected objects:")
        for box in results.boxes:
            cls_name = model.names[int(box.cls)]
            conf = float(box.conf)
            if conf >= conf_thresh:
                print(f"{cls_name} ({conf:.2f})")

    elif file_type == 'video':
        print("Processing video, this may take a while...")
        # YOLOv8 internal video processing (faster & more reliable)
        results = model.predict(file_path, save=True, conf=conf_thresh)
        video_file = results[0].path
        print(f"Video processed! Saved at: {video_file}")

        # Collect detection data
        detections = []
        for frame_idx, frame_result in enumerate(results[0].frames):
            for box in frame_result.boxes:
                cls_name = model.names[int(box.cls)]
                conf = float(box.conf)
                if conf >= conf_thresh:
                    detections.append({
                        "frame": frame_idx+1,
                        "class": cls_name,
                        "confidence": conf
                    })

        df = pd.DataFrame(detections)
        df.to_csv("video_detections.csv", index=False)
        print("Detections saved in 'video_detections.csv'")
        display(Video(video_file, embed=True))

    else:
        print("Invalid file type")
        continue

    if input("\nDetect another file? (y/n): ").lower() != 'y':
        break
```
